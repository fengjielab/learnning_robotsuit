```python
import robosuite as suite
import numpy as np
import imageio
config = {
    "env_name": "PickPlaceCans",
    "robots": "Panda",
    "render_camera": "frontview",
    "has_renderer": False,
    "use_camera_obs": True,
    "has_offscreen_renderer": True,
    "camera_heights": 1024,
    "camera_widths": 1024
}
```

```python
for i in range(20):
    env = suite.make(
        **config
    )
    env.sim.forward()
    fig = env.sim.render(
        camera_name="frontview",
        height=1024,
        width=1024,
    )[
        ::-1, :, :
    ].astype(np.uint8)
    with imageio.get_writer(
        f"data/initial_images/{i}.png",
    ) as writer:
        writer.append_data(fig)
```
